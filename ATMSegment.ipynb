### ATM Transection Analysis 

```python
import numpy as np
import pandas as pd
from pandas import plotting

# for visualizations
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')

# for interactive visualizations
import plotly.offline as py
from plotly.offline import init_notebook_mode, iplot
import plotly.graph_objs as go
from plotly import tools
init_notebook_mode(connected = True)
import plotly.figure_factory as ff

# for path
import os
```

```python
data = pd.read_csv('dataset/atm_transaction.csv')
data.head()
```

```python
data[["year", "month", "day"]] = data["Date"].str.split("-", expand = True)
```

```python
data['ATM']=data['No.ATM'].str.slice(start=3)
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data=data.drop(['No.ATM','Date'],axis=1)
```

```python
data.info()
```

```python
# checking if there is any NULL data

data.isnull().any().any()
```

```python
import warnings
warnings.filterwarnings('ignore')

plt.rcParams['figure.figsize'] = (18, 8)

plt.subplot(1, 2, 1)
sns.set(style = 'whitegrid')
sns.distplot(data['Withdrawals'])
plt.title('Distribution of Annual Income', fontsize = 20)
plt.xlabel('Range of Annual Withdrawals')
plt.ylabel('Count')


plt.subplot(1, 2, 2)
sns.set(style = 'whitegrid')
sns.distplot(data['ATM'], color = 'red')
plt.title('Distribution of ATM', fontsize = 20)
plt.xlabel('Range of ATM')
plt.ylabel('Count')
plt.show()
```

```python
data['month'].unique()
```

```python
data.month = data.month.replace({'01':"January", "02":"February", "03":"March"})
```

```python
data['day'].unique()
```

```python
labels = ['January','February','March']
size = data['month'].value_counts()
colors = ['red', 'orange','green']
explode = [0,0,0]

plt.rcParams['figure.figsize'] = (9, 9)
plt.pie(size, colors = colors, explode = explode, labels = labels, shadow = True, autopct = '%.3f%%')
plt.title('Month', fontsize = 20)
plt.axis('off')
plt.legend()
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (15, 8)
sns.countplot(data['month'], palette = 'hsv')
plt.title('Distribution of Month', fontsize = 20)
plt.show()
```

```python
data.day = pd.to_numeric(data.day, errors='coerce').astype('Int64')
```

```python
data.year = pd.to_numeric(data.year, errors='coerce').astype('Int64')
```

```python
sns.pairplot(data)
plt.title('Pairplot for the Data', fontsize = 20)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (15, 8)
sns.heatmap(data.corr(), cmap = 'copper', annot = True)
plt.title('Heatmap for the Data', fontsize = 20)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (18, 7)
sns.boxenplot(data['month'], data['Withdrawals'], palette = 'Blues')
plt.title('Month vs Withdrawals, fontsize = 20')
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (18, 7)
sns.boxenplot(data['day'], data['Withdrawals'], palette = 'Blues')
plt.title('Day vs Withdrawals, fontsize = 20')
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (18, 7)
sns.boxenplot(data['year'], data['Withdrawals'], palette = 'Blues')
plt.title('Year vs Withdrawals, fontsize = 20')
plt.show()
```

```python
x = data['Withdrawals']
y = data['month']
z = data['day']

sns.lineplot(x, y, color = 'blue')
sns.lineplot(x, z, color = 'pink')
plt.title('Annual Income vs Age and Spending Score', fontsize = 20)
plt.show()
```
